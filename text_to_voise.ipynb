```python
import matplotlib.pyplot as plt
import numpy as np

# Forward kinematics function
def forward_kinematics(theta1, theta2, l1, l2):
    # Calculate the end effector (x, y) position using joint angles and link lengths
    x = l1 * np.cos(theta1) + l2 * np.cos(theta1 + theta2)
    y = l1 * np.sin(theta1) + l2 * np.sin(theta1 + theta2)
    return x, y

# Inverse kinematics function
def inverse_kinematics(x, y, l1, l2):
    # Calculate the cosine of theta2 using the law of cosines
    D = (x**2 + y**2 - l1**2 - l2**2) / (2 * l1 * l2)
    # Calculate theta2 using the arctan2 function and trigonometric identities
    theta2 = np.arctan2(np.sqrt(1 - D**2), D)
    # Calculate theta1 using the arctan2 function and trigonometric identities
    theta1 = np.arctan2(y, x) - np.arctan2(l2 * np.sin(theta2), l1 + l2 * np.cos(theta2))
    return theta1, theta2

# Robot parameters
l1 = 1.0
l2 = 0.8

# Forward kinematics
theta1 = np.pi/4
theta2 = np.pi/3
x, y = forward_kinematics(theta1, theta2, l1, l2)
print("Forward Kinematics - End Effector Position: (", x, ",", y, ")")

# Inverse kinematics
x_target = 1.5
y_target = 0.3
theta1, theta2 = inverse_kinematics(x_target, y_target, l1, l2)
print("Inverse Kinematics - Joint Angles: (", theta1, ",", theta2, ")")

# Plot the robot motion
fig, ax = plt.subplots()
ax.plot([0, l1*np.cos(theta1), x_target], [0, l1*np.sin(theta1), y_target], 'o-')
ax.plot([l1*np.cos(theta1), x_target], [l1*np.sin(theta1), y_target], 'ro--')
ax.set_xlim([-1.5, 2.5])
ax.set_ylim([-1.5, 2.5])
ax.set_aspect('equal')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Robot Motion')
plt.show()
```
